# Data Preparation

```python
# show data stats and check duplications
print("Informasi Dataset:")
data.info()

print("\nStatistik Deskriptif:")
print(data.describe())

duplicates = data.duplicated().sum()
print(f"\nJumlah duplikat: {duplicates}")
```

```python
# date conversion to datetime
data['track_album_release_date'] = pd.to_datetime(data['track_album_release_date'], errors='coerce')
```

```python
# change duration metric which was originally from "ms" to "s"
data['duration_sec'] = data['duration_ms'] / 1000
data.head()
```

```python
# drop duration_ms
data = data.drop('duration_ms', axis=1)
data.head()
```

```python
# check missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# drop data with missing values
data_cleaned = data.dropna()

# show data before n after
print(f"Jumlah data sebelum menghapus missing values: {len(data)}")
print(f"Jumlah data setelah menghapus missing values: {len(data_cleaned)}")

# show data after removing missing values
print("\nData setelah menghapus missing values:")
data_cleaned.head()
```

```python
# merge columns for analysis
# by combining genre dan subgenre for max result
data_cleaned['playlist_full_genre'] = data_cleaned['playlist_genre'] + ' - ' + data_cleaned['playlist_subgenre']

# show
print("\nData setelah menambahkan kolom 'playlist_full_genre':")
data_cleaned[['playlist_genre', 'playlist_subgenre', 'playlist_full_genre']].head()
```

```python
data_cleaned.head()
```

```python
# checking album age based on release date - now (in yrs)
data_cleaned['album_age'] = (pd.to_datetime('today') - data_cleaned['track_album_release_date']).dt.days / 365

# show
print("\nData setelah menambahkan fitur 'album_age' (umur album):")
data_cleaned[['track_album_release_date', 'album_age']].head()
```

```python
# choosing relevant features for analysis
features = ['track_name', 'track_artist', 'track_popularity', 'danceability', 'energy', 'tempo', 'playlist_full_genre', 'album_age', 'duration_sec']
data_cleaned = data_cleaned[features]

# show
print("\nData setelah memilih fitur relevan:")
data_cleaned.head()
```

# EDA and Visualizations

```python
# decide tresshold for popularity in order to reach an 'underrated'
# count mean of popularity
average_popularity = data_cleaned['track_popularity'].mean()
print(f"Rata-rata Popularitas Lagu: {average_popularity}")

# filter songs based on below mean popularity
underrated_songs = data_cleaned[data_cleaned['track_popularity'] < average_popularity]
print(f"Jumlah Lagu Underrated: {len(underrated_songs)}")
```

```python
# to fill out "chill and calm" contexts, im gonna first visualize 3 features
# danceability, energy, and tempo
import seaborn as sns
import matplotlib.pyplot as plt
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Danceability
sns.histplot(data_cleaned['danceability'], kde=True, color='skyblue', ax=axs[0])
axs[0].set_title('Distribusi Danceability')
axs[0].set_xlabel('Danceability')
axs[0].set_ylabel('Frequency')

# Energy
sns.histplot(data_cleaned['energy'], kde=True, color='orange', ax=axs[1])
axs[1].set_title('Distribusi Energy')
axs[1].set_xlabel('Energy')
axs[1].set_ylabel('Frequency')

# Tempo
sns.histplot(data_cleaned['tempo'], kde=True, color='green', ax=axs[2])
axs[2].set_title('Distribusi Tempo')
axs[2].set_xlabel('Tempo (BPM)')
axs[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Tentukan Kriteria Chill dan Calm yang saya inginkan, dimana :
1. Danceability yang cukup tinggi (> 0.5)
2. Energy yang cukup rendah (<0.4)
3. Tempo lebih rendah (<100 BPM)

```python
# Filter
chill_calm_songs = data_cleaned[(data_cleaned['danceability'] > 0.5) &
                                 (data_cleaned['energy'] < 0.4) &
                                 (data_cleaned['tempo'] < 100) &
                                 (data_cleaned['track_popularity'] < average_popularity)]

print(f"Jumlah Lagu Chill dan Calm: {len(chill_calm_songs)}")
```
